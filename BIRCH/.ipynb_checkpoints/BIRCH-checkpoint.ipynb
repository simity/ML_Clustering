```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import Birch
clusModel=Birch(threshold=1,branching_factor=1000)
label=clusModel.fit_predict(x)
supervised=pd.DataFrame(dataset)
supervised["Cluster Group"]=label
```

```python
np.unique(label)
```

```python
supervised
```

```python
supervised.to_csv("Cluster.csv",index=False)
```

```python
import seaborn as sns
pict=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],legend=True,legend_out=True,fit_reg=False)
```
