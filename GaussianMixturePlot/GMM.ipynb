```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
x=pd.DataFrame(x)
```

```python
x
```

```python
from sklearn.mixture import GaussianMixture
clusModel=GaussianMixture(n_components=5, random_state=0)
label=clusModel.fit(x)
```

```python
label=label.fit_predict(x)
supervised=pd.DataFrame(dataset)
supervised["Cluster Group"]=label
```

```python
supervised
```

```python
import seaborn as sns
pict=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
