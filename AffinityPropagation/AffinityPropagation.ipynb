```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import AffinityPropagation
clusModel=AffinityPropagation(damping=0.9,max_iter=200)
label=clusModel.fit_predict(x)
supervised=pd.DataFrame(dataset)
```

```python
supervised["Cluster Group"]=label
supervised
```

```python
np.unique(label)
```

```python
import seaborn as sns
pict=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
