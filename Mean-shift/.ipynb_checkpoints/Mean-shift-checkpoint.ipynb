```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values 
```

```python
x
```

```python
from sklearn.cluster import MeanShift
clusModel=MeanShift(bandwidth=1)
label=clusModel.fit(x)
label=label.fit_predict(x)
```

```python
supervised=pd.DataFrame(dataset)
supervised["Cluster Group"]=label
```

```python
supervised
```

```python
supervised.to_csv("Cluster.csv", index=False)
```

```python
import seaborn as sns
pict=sns.lmplot(data=supervised,x= supervised.columns[3],y= supervised.columns[4], hue= supervised.columns[5],fit_reg=False,,legend=True,legend_out=False)
```
